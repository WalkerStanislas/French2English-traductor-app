# III- Cleaning dataset

```python
def clean_text(text):
    text = text.lower()
    text = text.strip()
    text = " ".join(text.split())
    return text
```

```python
src_clean = [clean_text(line) for line in src_lines]
tgt_clean = [clean_text(line) for line in tgt_lines]
for i in range(5):
    print(f"FR: {src_clean[i]}")
    print(f"EN: {tgt_clean[i]}")
    print("-" * 50)
```

# V- OpenNMT prepossing data step

## V.1- Tokenization of train and validation data

```python
# We shall use Spacy, then let load french and english tokenizer
import spacy
```

```python
!python -m spacy download fr_core_news_sm
!python -m spacy download en_core_web_sm
```

```python
nlp_fr = spacy.load("fr_core_news_sm")
nlp_en = spacy.load("en_core_web_sm")
```

```python
print("French model :", nlp_fr)
print("English model :", nlp_en)
```

```python
!pip install tqdm # Follow the process
```

```python
from tqdm import tqdm

# Tokenize per batch
def tokenize_batch(lines, nlp, batch_size=1000):
    tokenized_lines = []
    for doc in tqdm(nlp.pipe(lines, batch_size=batch_size), total=len(lines), desc="Tokenisation"):
        tokenized_lines.append(" ".join([token.text for token in doc]))
    return tokenized_lines
```

```python
src_val = tokenize_batch(src_val, nlp_fr)
tgt_val = tokenize_batch(tgt_val, nlp_en)
src_train = tokenize_batch(src_train, nlp_fr)
tgt_train = tokenize_batch(tgt_train, nlp_en)
```

```python
def save_data(filename, data):
    with open(filename, "w", encoding="utf-8") as f:
        f.writelines(line + "\n" for line in data)
```

```python
save_data("./data/fr_train_tokenized.txt", src_train)
save_data("./data/en_train_tokenized.txt", tgt_train)
save_data("./data/fr_val_tokenized.txt", src_val)
save_data("./data/en_val_tokenized.txt", tgt_val)
```
